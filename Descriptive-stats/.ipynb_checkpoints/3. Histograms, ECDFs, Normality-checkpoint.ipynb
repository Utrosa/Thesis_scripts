### Empirical cumulative distribution function (ECDF)

```python
# Plots all of the data (no bining bias)
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)
    # x-data for the ECDF: x
    x = np.sort(data)
    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n
    return x, y
```

```python
# Generate ECDF for f0
p = [111, 555, 777, 888, 999]
for participant in p:
    x_obs, y_obs = ecdf(df[df['PARTICIPANT_ID'] == participant]['FUNDAMENTAL_FREQUENCY'])
    plt.plot(x_obs, y_obs, marker = '.')
    plt.xlabel('Fundamental Frequency')
    plt.ylabel('ECDF')
    plt.legend(('111', '555', '777', '888', '999'), loc='lower right')
    plt.title('Empirical Cumulative Distribution Function - Pilot')
    
#plt.savefig('ECDF-SF-f0.png', dpi=1800) # Change the filename
```

```python
# Generate ECDF for pitch
p = [111, 555, 777, 888, 999]
for participant in p:
    x_obs, y_obs = ecdf(df[df['PARTICIPANT_ID'] == participant]['PITCH'])
    plt.plot(x_obs, y_obs, marker = '.', linestyle = 'none')
    plt.xlabel('Hz')
    plt.ylabel('ECDF')
    plt.legend(('111', '555', '777', '888', '999'), loc='lower right')
    plt.title('Empirical Cumulative Distribution Function - Pilot')
    
#plt.savefig('ECDF-SF-Hz.png', dpi=1800) # Change the filename
```

##### Comparing data to a Normal Probability Density Funtion (PDF) 

##### f0

```python
# Check for normality of f0 distribution

p = [111, 555, 777, 888, 999]
obs = []
theor = []
for participant in p:
    mean = np.mean(df[df['PARTICIPANT_ID'] == participant]['FUNDAMENTAL_FREQUENCY'])
    std = np.std(df[df['PARTICIPANT_ID'] == participant]['FUNDAMENTAL_FREQUENCY'])

    samples = np.random.normal(mean, std, size=len(df[df['PARTICIPANT_ID'] == participant]['FUNDAMENTAL_FREQUENCY']))
    x_obs, y_obs = ecdf(df[df['PARTICIPANT_ID'] == participant]['FUNDAMENTAL_FREQUENCY'])
    x_theor, y_theor = ecdf(samples)
    obs.append([x_obs, y_obs])
    theor.append([x_theor, y_theor])

# Plot the theoretical and empirical (observed) CDF per participant

fig, ((ax0, ax1, ax2), (ax3, ax4, ax5)) = plt.subplots(nrows=2, ncols=3)

_ = ax0.plot(theor[0][0], theor[0][1])
_ = ax0.plot(obs[0][0], obs[0][1], marker='.', linestyle='none', color='r')
_ = ax0.set_xlabel('Fundamental Frequency')
_ = ax0.set_ylabel('CDF')
_ = ax0.set_title('111')

_ = ax1.plot(theor[1][0], theor[1][1])
_ = ax1.plot(obs[1][0], obs[1][1], marker='.', linestyle='none', color='r')
_ = ax1.set_xlabel('Fundamental Frequency')
_ = ax1.set_ylabel('CDF')
_ = ax1.set_title('555')

_ = ax2.plot(theor[2][0], theor[2][1])
_ = ax2.plot(obs[2][0], obs[2][1], marker='.', linestyle='none', color='r')
_ = ax2.set_xlabel('Fundamental Frequency')
_ = ax2.set_ylabel('CDF')
_ = ax2.set_title('777')

_ = ax3.plot(theor[3][0], theor[3][1])
_ = ax3.plot(obs[3][0], obs[3][1], marker='.', linestyle='none', color='r')
_ = ax3.set_xlabel('Fundamental Frequency')
_ = ax3.set_ylabel('CDF')
_ = ax3.set_title('888')

_ = ax4.plot(theor[4][0], theor[4][1])
_ = ax4.plot(obs[4][0], obs[4][1], marker='.', linestyle='none', color='r')
_ = ax4.set_xlabel('Fundamental Frequency')
_ = ax4.set_ylabel('CDF')
_ = ax4.set_title('999')

ax5.set_visible(False)

fig.tight_layout()
fig.suptitle('f0 Normality for Sung Improvisations - Pilot Data') # Change the title
fig.set_figheight(10)
fig.set_figwidth(15)
fig.suptitle('f0 Normality - Pilot Data', ha ='center', fontsize=20)

fig.savefig("Normality-f0-Improv.png", dpi=2000) # Change the filename
```
