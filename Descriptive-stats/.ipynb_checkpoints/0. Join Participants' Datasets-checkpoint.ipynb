## Join the datasets of participants.

### Import packages.

```python
import pandas as pd
import os, sys
```

### Load spectral datasets and join them.

```python
# Specify the paths to the dataframes.
condition = 'IM' # IM or SF
directory = f"C:/Users/monik/Documents/Paris/Internship/Data-Collection/Extracted-Data/Glides-Notes/{condition}/"
dirs = os.listdir(directory)

# Check the number of dataframes to import
print(len(dirs))

# Read in the dataframes
spectral_data = pd.DataFrame()
for filepath in dirs:
    df = pd.read_csv(f'{directory}{filepath}', index_col=0)
    spectral_data = pd.concat([spectral_data, df], ignore_index=True, axis=0)
```

```python
# Check the data.
spectral_data.head()
```

```python
# Check participants & Song IDs
print(spectral_data['PARTICIPANT_ID'].unique())
print(spectral_data['SONG_ID'].unique())
```

## Export the datasets.

```python
# Save to csv.
spectral_data.to_csv(f"{directory}JOINED_data_{condition}.csv")
```
