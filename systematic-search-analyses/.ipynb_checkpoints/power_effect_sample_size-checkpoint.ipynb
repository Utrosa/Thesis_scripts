# Power and Sample Size Calculation

To estimate the power and minimal sample size for detecting static functional connectivity (FC) alterations in Autism Spectrum Disorder (ASD) as well as correlation between static FC and clinical symptoms, we adapted the power demo code from Koen Helwegen (https://github.com/dutchconnectomelab/power-demo/blob/main/setup.py). The power demo accompanies their paper on statistical power in network neuroscience: https://doi-org.portail.psl.eu/10.1016/j.tics.2022.12.011.
The main adaptations are:
1. Removal of Case 2 for structural connectivity.
2. Using FDR method to determine the significance level (alpha) instead of Bonferroni.
3. Running the script for multiple edges .

```python
# User-defined variables
power_convention = 0.8
alpha_FDR = 0.05
rho = 0.8
mean_edge_weight_user = 0.53

# Set saving options: 1 = yes, 0 = no
save = 1
```

```python
# Import necessary packages and modules.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
from statsmodels.stats.power import tt_ind_solve_power
from statsmodels.stats import power as smpower
from pandas.core.computation.check import NUMEXPR_INSTALLED

import pinn

# Set pretty print
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 3)
```

#### Functions

```python
def tt_critical_effect_size(sample_size: int, alpha: float = 0.05) -> float:
    """Find the smallest effect size (Cohen's d) that is still significant in a two-tailed ttest."""
    return tt_ind_solve_power(nobs1=sample_size, alpha=alpha, power=power_convention, reliability=rho)

def tt_ind_solve_power(
    effect_size=None,
    nobs1=None,
    alpha=None,
    power=None,
    reliability=rho,
    **kwargs,
):
    """Wrapper of statsmodels.stats.power.tt_ind_solve_power with reliability correction"""
    if not (0 <= reliability <= 1):
        raise ValueError(
            f"Reliability score should lie in [0,1], instead got {reliability}."
        )

    if effect_size is not None:
        return smpower.tt_ind_solve_power( # Default is two-sided t-test. Outputs the parameter which is set to None.
            effect_size=effect_size * np.sqrt(reliability), # Correction for measurement error in the data
            nobs1=nobs1,
            alpha=alpha,
            power=power,
            **kwargs,
        )
    else:
        return smpower.tt_ind_solve_power(
            nobs1=nobs1,
            alpha=alpha,
            power=power,
            **kwargs,
        ) / np.sqrt(reliability)

# Algorithm to estimate power. It will return power if that's the only required parameter set to None.
def measurement_error():
    powers = []
    for ix, row in df_get_power.iterrows():

        # Extract values effect_size, sample_size, alpha, and reliability from the current row
        effect_size_data = row["cohen_d"]
        sample_size_data = row['sample_size']
        alpha_data = row['alpha']
        reliability_data = row['reliability']
#         print(f"Effect Size: {effect_size}, Sample Size: {sample_size}, Alpha: {alpha}, Reliability: {reliability}")

        #Call the function tt_ind_solve_power with the extracted values
        edge_power = tt_ind_solve_power(effect_size=effect_size_data, 
                                        nobs1=sample_size_data, 
                                        alpha=alpha_data, 
                                        power=None,
                                        reliability=reliability_data)
        powers.append(edge_power)
    return powers

# Perform power simulation based on effect size, reliability, and significance level.
def monte_carlo_power(
    mean_controls,
    mean_patients,
    sample_size,
    alpha,
    rm_zeros=None,
    iterations=10_000,
    edge_generator=None,
):
    """Estimate power for edgewise comparison with missing values.

    Args:
      mean_controls: mean edge weight for the control group
      mean_patients: mean edge weight for the patient group
      sample_size: number of observations per group
      alpha: significance threshold
      rm_zeros: whether or not to remove missing values before inference
      iterations: number of simulated studies used to estimate power
      edge_generator: function used to generate edges. Should accept arguments `sample_size` and `mean`

    Return power (ratio of studies in which null hypothesis was correctly rejected)
    """
    # The main loop, simulating studies while keeping
    # track of how many are successful.
    detection_count = 0
    for _ in range(iterations):
        edges_controls = edge_generator(sample_size=sample_size, mean=mean_controls)
        edges_patients = edge_generator(sample_size=sample_size, mean=mean_patients)

        if rm_zeros:
            edges_controls = edges_controls[edges_controls != 0]
            edges_patients = edges_patients[edges_patients != 0]

        # we count a detection as successful if it is significant (at alpha)
        # and in the right direction
        if (ttest_ind(edges_controls, edges_patients)[1] < alpha) and (
            np.sign(np.mean(edges_controls) - np.mean(edges_patients))
            == np.sign(mean_controls - mean_patients)
        ):
            detection_count += 1

    # our power estimation is simply the ratio of successful studies
    return detection_count / iterations

# Define a function to calculate sample sizes for different groups
def calculate_sample_sizes(effect_sizes, alpha, power, reliability):
    sample_sizes = {}
    for group, effect_size in effect_sizes.items():
        sample_sizes[group] = tt_ind_solve_power(effect_size=effect_size,
                                                 alpha=alpha,
                                                 power=power,
                                                 reliability=reliability)
    return sample_sizes
```

## Power Computation

Specify the required information for edge-level power computation: 
- effect size, 
- significance level,
- power, and 
- degree of measurement error (rho).

Define also the **ratio missing** (only relevant for section Other).

We'll use the FDR approach to correct for family-wise error rates, which means that alpha is set to 0.05. If we would use Bonferroni method, we would have to divide alpha with the number of multiple comparisons. The conventional power for neuroimaging studies is 0.8.

```python
# Specify the file with the required data collected from the systematic search
file = "Meta-Analysis-Static-FC.xlsx"

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(file, sheet_name='data')  # Change 'data' to the actual sheet name if it's different

# Look
display(df)
```

```python
# Extract only the columns needed for power computation and plotting
columns = ["paper_shortname","ABIDE", "age_cat", "age", "effect_type", "cohen_d",
           "sample_size", "alpha", "power", "reliability"]

df_get_power = df[columns]

# Turn Cohen's d into float
df_get_power.loc[:, 'cohen_d'] = df_get_power['cohen_d'].astype(float)

# Look
display(df_get_power)
```

Calculate the power of studies that report effect sizes of static functional connectivity alterations in Autism Spectrum Disorder (ASD). The studies considered are based on a systematic search of the literature.

```python
power_result = measurement_error()

# Add the calculated power to the dataframe
df_get_power.loc[:, "power"] = power_result

# Look
display(df_get_power)
```

## Sample Size Computation

```python
# Calculate the min. effect size given a sample size
max_sample = 90
min_effect_size = tt_critical_effect_size(sample_size=max_sample)

print(f"The minimum effect size for sample size {max_sample} is {round(min_effect_size, 2)}.")
```

```python
# Run the algorithm to estimate sample size. It returns sample size (nobs1) because it's the only parameter set to None.
# Note that the estimate of the sample size is per group for a two-sample independent t-test design.

# Define the data
ex = {
    "all records (all)": mean_effect_all,
    "all records (static FC)": mean_effect_static_FC_all,
    "all records (pheno)": mean_effect_pheno_all,
    
    "adults (all)": mean_effect_all_adults,
    "adults (static FC)": mean_effect_static_FC_adults,
    "adults (pheno)": mean_effect_pheno_adults,
    
    "children (all)": mean_effect_all_children,
    "children (static FC)": mean_effect_static_FC_children,
    "children (pheno)": mean_effect_pheno_children
}

# Calculate sample sizes for each effect type
sample_size_results = calculate_sample_sizes(ex, alpha_FDR, power_convention, rho)

# Create DataFrame
df_summary_sample_sizes = pd.DataFrame({
    "Reliability": rho,
    "Power": power_convention,
    "Mean Effect Size": list(ex.values()),
    "Min Sample Size": [round(sample_size) for sample_size in sample_size_results.values()]
}, index=ex.keys())

# Save DataFrame to CSV
if save == 1:
    df_summary_sample_sizes.to_csv('power-results/summary_sample_sizes.csv')

# Look
display(df_summary_sample_sizes)
```

```python
# Get reliabilities 
reliabilities = np.arange(0.1, 1.1, 0.1)

samplesize_ALL_ad = []
samplesize_STATIC_FC_ad = []
samplesize_STATIC_FC_SYMPTOMS_ad = []

samplesize_ALL_ch = []
samplesize_STATIC_FC_ch = []
samplesize_STATIC_FC_SYMPTOMS_ch = []

for r in reliabilities: 
    # Adults
    samplesize_all_ad = tt_ind_solve_power(effect_size=mean_effect_all,alpha=alpha_FDR,power=power_convention,
                                        reliability=r)
    samplesize_static_FC_ad = tt_ind_solve_power(effect_size=mean_effect_static_FC_all,alpha=alpha_FDR,power=power_convention,
                                              reliability=r)
    samplesize_static_FC_symptoms_ad = tt_ind_solve_power(effect_size=mean_effect_pheno_all,
                                                       alpha=alpha_FDR,power=power_convention, reliability=r)
    # Children
    samplesize_all_children = tt_ind_solve_power(effect_size=mean_effect_all_children,alpha=alpha_FDR,power=power_convention,
                                        reliability=r)
    samplesize_static_FC_children = tt_ind_solve_power(effect_size=mean_effect_static_FC_children,alpha=alpha_FDR,
                                                       power=power_convention,reliability=r)
    samplesize_static_FC_symptoms_children = tt_ind_solve_power(effect_size=mean_effect_pheno_children,
                                                       alpha=alpha_FDR,power=power_convention, reliability=r)
    # Append
    samplesize_ALL_ad.append(samplesize_all_ad)
    samplesize_STATIC_FC_ad.append(samplesize_static_FC_ad)
    samplesize_STATIC_FC_SYMPTOMS_ad.append(samplesize_static_FC_symptoms_ad)
    
    samplesize_ALL_ch.append(samplesize_all_children)
    samplesize_STATIC_FC_ch.append(samplesize_static_FC_children)
    samplesize_STATIC_FC_SYMPTOMS_ch.append(samplesize_static_FC_symptoms_children)
```

```python
# Plot sample sizes over reliabilities for adults and children
sns.set_theme(style='whitegrid')
fig, axs = plt.subplots(1, 2, figsize=(20,10), sharex=True)

axs[0].plot(reliabilities, samplesize_ALL_ad, color='#00ba77', label="All group difference effects")
axs[0].plot(reliabilities, samplesize_STATIC_FC_ad, color='#A500FF', label="Static FC differences")
axs[0].plot(reliabilities, samplesize_STATIC_FC_SYMPTOMS_ad, color='#ffa500', 
            label="Correlation between static FC and clinical symptoms")
axs[0].tick_params(axis='x', labelsize=14)
axs[0].tick_params(axis='y', labelsize=14)
axs[0].set_ylabel("Sample Size per Group", fontsize=18, labelpad=12)
axs[0].set_ylim([0, 200])
axs[0].set_title("Adults (18+)", fontsize=20, y=1.02)

axs[1].plot(reliabilities, samplesize_ALL_ch, color='#00ba77', label="All group difference effects")
axs[1].plot(reliabilities, samplesize_STATIC_FC_ch, color='#A500FF', label="Static FC differences")
axs[1].plot(reliabilities, samplesize_STATIC_FC_SYMPTOMS_ch, color='#ffa500', 
            label="Correlation between static FC and clinical symptoms")
axs[1].tick_params(axis='x', labelsize=14)
axs[1].tick_params(axis='y', labelsize=14)
axs[1].set_ylabel("Sample Size per Group", fontsize=18, labelpad=12)
axs[1].set_ylim([0, 200])
axs[1].set_title("Children (6 - 13)", fontsize=20, y=1.02)

for ax in axs:
    ax.set_xlabel("Reliability", fontsize=18, labelpad=12)

fig.suptitle("Minimal Sample Size Depends on Reliability (CH)", fontsize=22, y=1.08)
handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', fontsize='large', bbox_to_anchor=(0.5, 1.03), ncol=3)

if save == 1:
    plt.savefig("power-results/min-sample-size-over-reliabilities.png",bbox_inches='tight')
    
plt.show()
```
