```python
import pandas as pd
import numpy as np
from scipy.spatial import distance
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/monik/Documents/Paris/Internship/Analyses/Inferential-stats/777_ew_insp_spectral.csv")
data = pd.read_csv("C:/Users/monik/Documents/Paris/Internship/Analyses/Pilot-Data/Keith-2-Data/spectral_data_777_IMPROV.csv")
```

```python
# Get probability distribution of one song
df['SONG_NAME'].unique()
```

```python
song1 = df[df['SONG_NAME'] == 'KillBill']['MIDI_NOTE'].to_numpy()
```

```python
jsd = []

for idx in data.SONG_ID.unique():
    improv = data[data['SONG_ID'] == idx]['MIDI_NOTE'].to_numpy()
#     plt.hist(improv)
#     plt.hist(song1,alpha=0.2)
#     plt.show()
    
    n_bins = 128
    imp_freq, imp_bins = np.histogram(improv[~np.isnan(improv)],bins=n_bins)
    song_freq, song_bins = np.histogram(song1[~np.isnan(song1)],bins=n_bins)
    
#     print(imp_freq)
#     print(song_freq)
#     print(imp_freq.shape)
#     print(song_freq.shape)
    jsd.append(distance.jensenshannon(imp_freq,song_freq))

#     break
    
jsd = np.array(jsd)    

print(jsd)
print(jsd.max(),jsd.min())


plt.bar(np.arange(len(jsd)),jsd)
```
